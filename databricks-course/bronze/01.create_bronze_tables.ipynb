### Crear tablas para archivos CSV

#### Crear la tabla "movies"

cargo un archivo que se encuentra en una ubicacion externa hacia Databricks

```sql
DROP TABLE IF EXISTS movie_bronze.movies;
CREATE TABLE IF NOT EXISTS movie_bronze.movies (
    movieId INT,
    title STRING,
    budget DOUBLE,
    homePage STRING,
    overview STRING,
    popularity DOUBLE,
    yearReleaseDate INT,
    releaseDate DATE,
    revenue DOUBLE,
    durationTime INT,
    movieStatus STRING,
    tagline STRING,
    voteAverage DOUBLE,
    voteCount INT
)
USING CSV
OPTIONS (path "abfss://bronze@moviehistory22.dfs.core.windows.net/movie.csv", header true);
```

```sql
SELECT * FROM movie_bronze.movies;
```
