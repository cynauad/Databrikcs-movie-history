### Acceso al DataFrame mediante SQL

#### Local Temp View

    1. Crear la vista(view) temporal en la Base de Datos
    2. Acceder a lavista desde la celda "SQL"
    3. Acceder a la vista desde la celda "Python"
    4. Acceder a la vista desde otro Notebook

```python
%run "../includes/configuration"
```

```python
results_movie_genre_languages_df = spark.read.parquet(f"{gold_folder_path}/results_movie_genre_language")    
```

```python
# creacion de la vista
results_movie_genre_languages_df.createTempView("v_movie_genre_language")

# results_movie_genre_languages_df.createOrReplaceTempView("v_movie_genre_language")
```

```python
%sql
SELECT * 
FROM v_movie_genre_language
WHERE vote_average > 7.5
```

```python
display(results_movie_genre_languages_df)
```

```python
results_movie_genre_languages_2 = spark.sql("""SELECT * 
FROM v_movie_genre_language
WHERE vote_average > 7.5""")
```

```python
display(results_movie_genre_languages_2)
```

```python
vote_average = 7.5
```

```python
results_movie_genre_languages_2 = spark.sql(f"SELECT * FROM v_movie_genre_language WHERE vote_average > {vote_average}")
```

```python
display(results_movie_genre_languages_2)
```

### Global Temp View

#### Objetivos
    1. Crear la vista(view) temporal global del Dataframe
    2. Acceder a lavista desde la celda "SQL"
    3. Acceder a la vista desde la celda "Python"
    4. Acceder a la vista desde otro Notebook

```python
results_movie_genre_languages_df.createOrReplaceGlobalTempView("gv_movie_genre_language")
```

```python
%sql
SHOW TABLES IN global_temp;
```
