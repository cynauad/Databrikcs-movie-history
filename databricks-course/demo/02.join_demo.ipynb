### Spark Join Transformation

#### INNER JOIN

```python
%run "../includes/configuration"
```

```python
movies_df = spark.read.parquet(f"{silver_folder_path}/movies") #.filter(movie_df.year_release_date == 2007)
```

```python
production_country_df = spark.read.parquet(f"{silver_folder_path}/productions_countries")
```

```python
country_df = spark.read.parquet(f"{silver_folder_path}/countries") 
```

```python
display(movies_df)
```

```python
display(production_country_df)
```

```python
display(country_df)
```

```python
movie_production_country_df = movies_df.join(production_country_df, 
                                        movies_df.movie_id == production_country_df.movie_id, 
                                        "inner")  \
                                        .select(movies_df.movie_id, movies_df.title, movies_df.budget, production_country_df.country_id)

# inner es default, se puede omitir
```

```python
display(movie_production_country_df)
```

```python
movie_country_df = movie_production_country_df.join(country_df,
                                        movie_production_country_df.country_id == country_df.country_id,
                                        "inner") \
                                        .select(movie_production_country_df.title, movie_production_country_df.budget, country_df.country_name)
```

```python
display(movie_country_df)
```

#### OUTER JOIN

##### Left Outer Join

```python
movie_production_country_df = movies_df.join(production_country_df, 
                                        movies_df.movie_id == production_country_df.movie_id, 
                                        "left")  \
                                        .select(movies_df.movie_id, movies_df.title, movies_df.budget, production_country_df.country_id)
```

```python
display(movie_production_country_df)
```

```python
movie_country_df = movie_production_country_df.join(country_df,
                                        movie_production_country_df.country_id == country_df.country_id,
                                        "left") \
                                        .select(movie_production_country_df.title, movie_production_country_df.budget, country_df.country_name)
```

```python
display(movie_country_df)
```

##### Right Outer join

```python
movie_production_country_df = movies_df.join(production_country_df, 
                                        movies_df.movie_id == production_country_df.movie_id, 
                                        "right")  \
                                        .select(movies_df.movie_id, movies_df.title, movies_df.budget, production_country_df.country_id)
```
