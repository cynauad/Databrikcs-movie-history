### Acceder a Azure Data Lake Storage mediante Service Principal
1. "Registrar la Aplicación" en Azure Entra ID/ Service Principal
2. Generar un secreto (contraseña) para la aplicación
3. Configurar Spark con APP/Client Id, Directory/ Tenand Id & Secret
4. Asignar el Role "Storage Blob data Contributor" al Data Lake

```python
dbutils.secrets.listScopes()
```

```python
dbutils.secrets.list("movie-history-secret-sope")
```

```python
dbutils.secrets.get(scope="movie-history-secret-sope", key="movie-client-id")
dbutils.secrets.get(scope="movie-history-secret-sope", key="movie-tenant-id")
dbutils.secrets.get(scope="movie-history-secret-sope", key="movie-client-secret")
```

```python
client_id = dbutils.secrets.get(scope="movie-history-secret-sope", key="movie-client-id")
tenant_id = dbutils.secrets.get(scope="movie-history-secret-sope", key="movie-tenant-id")
client_secret = dbutils.secrets.get(scope="movie-history-secret-sope", key="movie-client-secret")
```

```python
spark.conf.set("fs.azure.account.auth.type.moviehistory22.dfs.core.windows.net", "OAuth") # autenticacion mediante OAuth
spark.conf.set("fs.azure.account.oauth.provider.type.moviehistory22.dfs.core.windows.net", 
               "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")         
spark.conf.set("fs.azure.account.oauth2.client.id.moviehistory22.dfs.core.windows.net", client_id) 
spark.conf.set("fs.azure.account.oauth2.client.secret.moviehistory22.dfs.core.windows.net", client_secret) 
spark.conf.set("fs.azure.account.oauth2.client.endpoint.moviehistory22.dfs.core.windows.net", 
               F"https://login.microsoftonline.com/{tenant_id}/oauth2/token")             
```

```python
display(dbutils.fs.ls("abfss://demo@moviehistory22.dfs.core.windows.net"))
```
