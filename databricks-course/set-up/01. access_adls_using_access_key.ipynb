### Acceder a Azure Data Lake Storage mediante Access Key
1. Establecer la configuración de spark "fs.azure.account.key"
2. listar archivos del contenedor "demo"
3. leer datos del archivo "movie.csv"

```python
# esto lo agrego despues de crear el secreto.

movie_access_key = dbutils.secrets.get(scope = "movie-history-secret-sope", key = "movie-access-key")
```

```python
spark.conf.set(
    "fs.azure.account.key.moviehistory22.dfs.core.windows.net",
    movie_access_key
)
```

```python
display(dbutils.fs.ls("abfss://demo@moviehistory22.dfs.core.windows.net"))
```

```python
display(spark.read.csv("abfss://demo@moviehistory22.dfs.core.windows.net/movie.csv"))
```
