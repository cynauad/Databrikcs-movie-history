### Acceder a Azure Data Lake Storage mediante Token Sas
1. Establecer la configuración de spark "SAs Token"
2. listar archivos del contenedor "demo"
3. leer datos del archivo "movie.csv"

OJO!!! las SAS token expiran!!! se debe crear una nueva para probaar nuevamente!!!!

```python
dbutils.secrets.listScopes()
```

```python
dbutils.secrets.list(scope = "movie-history-secret-sope")
```

```python
movie_token_sas = dbutils.secrets.get(scope = "movie-history-secret-sope", key = "movie-sas-token")
```

```python
spark.conf.set("fs.azure.account.auth.type.moviehistory22.dfs.core.windows.net", "SAS") # autenticacion mediante sas
spark.conf.set("fs.azure.sas.token.provider.type.moviehistory22.dfs.core.windows.net", 
               "org.apache.hadoop.fs.azurebfs.sas.FixedSASTokenProvider")          # conexion mediante provedor token sas
spark.conf.set("fs.azure.sas.fixed.token.moviehistory22.dfs.core.windows.net", 
               movie_token_sas)  # especificar conexion
```

```python
display(dbutils.fs.ls("abfss://demo@moviehistory22.dfs.core.windows.net"))
```

```python
display(spark.read.csv("abfss://demo@moviehistory22.dfs.core.windows.net/movie.csv"))
```
